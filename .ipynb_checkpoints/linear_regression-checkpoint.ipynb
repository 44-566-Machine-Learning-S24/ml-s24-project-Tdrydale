## Part 1 - Initial data prep section.  Read, clean and create sets.

```python
import pandas as pd

data_frame = pd.read_csv("Housing_data.gitignored/realtor-data.csv")
data_frame = data_frame.dropna(thresh=5)
filtered_data = data_frame[data_frame["bed"] < 10]
filtered_data = filtered_data[filtered_data["bath"] < 10]
filtered_data = filtered_data[filtered_data["acre_lot"] < 3]
filtered_data = filtered_data[filtered_data["house_size"] < 5000]
filtered_data = filtered_data[filtered_data["price"] < 2000000]
filtered_data = filtered_data[filtered_data["state"] == "Delaware"]
```

```python
import numpy as np
def fractional_split(filtered_data, test_fraction=0.2, seed=42):
    data_count = len(filtered_data)
    test_count = int(test_fraction*data_count)
    
    np.random.seed(seed)
    shuffled_indices = np.random.permutation(data_count)
    
    test_indices = shuffled_indices[:test_count]
    train_indices = shuffled_indices[test_count:]
    
    return filtered_data.iloc[train_indices], filtered_data.iloc[test_indices]

train_set, test_set = fractional_split(filtered_data)
```

## Part 2 - Do a linear regression on X and y.  Compute metrics.

```python
working_set = train_set.copy()
import matplotlib.pyplot as plt
plt.scatter(working_set["house_size"], working_set["price"])
plt.xlabel("House size (sq ft)")
plt.ylabel("Price ($)")
plt.show()
```

I chose the house size and price because the house size has the highest correlation compaired to the price

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
X = working_set[["house_size"]]
Y = working_set["price"]
reg.fit(X, Y)
```

```python
print("The bias is " , reg.intercept_)
print("The feature coefficients are ", reg.coef_)
```

```python
theta0 = reg.intercept_
theta1 = reg.coef_[0]
print(theta0, theta1)
def the_line(h, bias, coef):
    return bias + coef * h

min_house_size = working_set["house_size"].min()
min_price = the_line(min_house_size, theta0, theta1)
max_house_size = working_set["house_size"].max()
max_price = the_line(max_house_size, theta0, theta1)
line_x = [min_house_size, max_house_size]
line_y = [min_price, max_price]


import matplotlib.pyplot as plt
plt.scatter(working_set["house_size"], working_set["price"])
plt.xlabel("House size (sq ft)")
plt.ylabel("price ($)")
plt.plot(line_x, line_y)
plt.show()
```

The line looks pretty good, it goes through the middle of all the data points

```python
print(reg.score(X,Y))
```

```python
from sklearn.metrics import mean_squared_error
y_pred = reg.predict(X)
mse = mean_squared_error(Y, y_pred)
print("mean squared error is ",  mse)

from math import sqrt
rmse = sqrt(mse)
print("The root mean squared error is ", rmse)
```

```python
band = 3*rmse

theta0 = reg.intercept_
theta1 = reg.coef_[0]
print(theta0, theta1)
def the_line(h, bias, coef):
    return bias + coef * h

min_house_size = working_set["house_size"].min()
min_price = the_line(min_house_size, theta0, theta1)
max_house_size = working_set["house_size"].max()
max_price = the_line(max_house_size, theta0, theta1)
line_x = [min_house_size, max_house_size]
line_y = [min_price, max_price]
upper_line_y = [min_price+band, max_price+band]
lower_line_y = [min_price-band, max_price-band]


import matplotlib.pyplot as plt
plt.scatter(working_set["house_size"], working_set["price"])
plt.xlabel("House size (sq ft)")
plt.ylabel("Price ($)")
plt.plot(line_x, upper_line_y)
plt.plot(line_x, line_y)
plt.plot(line_x, lower_line_y)
plt.show()
```

## Part 4 - See if you can do better. Do a final evaluation with the test set.

```python
from sklearn.linear_model import LinearRegression
reg_plus_cube = LinearRegression()

X = working_set[["house_size"]]
Y = working_set["price"]

from sklearn.preprocessing import PolynomialFeatures
create_cube_terms = PolynomialFeatures(degree=3, include_bias=False)
X_cube_terms = create_cube_terms.fit_transform(X)

                
                
reg_plus_cube.fit(X_cube_terms, Y)

print(reg_plus_cube.score(X_cube_terms,Y))

min_house_size = working_set["house_size"].min()
max_house_size = working_set["house_size"].max()
points = 1000
delta = (max_house_size-min_house_size)/points
cube_x = [min_house_size + i*delta for i in range(points)]
inputs = [[h, h*h, h**3] for h in cube_x]
cube_y = reg_plus_cube.predict(inputs)

import matplotlib.pyplot as plt
plt.scatter(working_set["house_size"], working_set["price"])
plt.xlabel("House Size (sqw ft)")
plt.ylabel("Price ($)")
plt.plot(cube_x, cube_y, color="red")
plt.show()
```

The price tends to go down/levels out when the house size grows, it's weird that the price would go down the bigger the house is, maybe the houses that were that big that sold did not have as much furniture or were in worse shape than some other houses
